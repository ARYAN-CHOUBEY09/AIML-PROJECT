```python
import pandas as pd
```

```python
df = pd.read_csv('FAISS_text.csv')
```

```python
#converting the text to vectors using embeddings
```

```python
from sentence_transformers import SentenceTransformer
```

```python
#using hugging face encoder 
encoder =  SentenceTransformer('all-mpnet-base-v2')
vectors = encoder.encode(df.text)
```

```python
vectors
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
dim  = vectors.shape[1]
```

```python
dim
```

```python
import faiss
index =  faiss.IndexFlatL2(dim)
```

```python
index
```

```python
index.add(vectors)
```

```python
search_query = "An apple a day keeps doctors away"
```
